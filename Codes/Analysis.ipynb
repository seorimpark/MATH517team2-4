```python
b=pd.DataFrame(data2.model.value_counts())
a=b[b["model"]>22]
c=a.to_dict()
d=c["model"]
plt.figure(figsize=(20,8))
plt.bar(list(d.keys()),np.array(list(d.values())))
plt.title("Major aircraft model names from which accidents occured")
plt.show()
```

```python
manu=pd.DataFrame(data2.manufacturer.value_counts())
manu=manu[manu.manufacturer>10]
manud=manu.to_dict()
manudr=manud["manufacturer"]
plt.figure(figsize=(20,8))
plt.bar(manudr.keys(),manudr.values())
plt.title("Major manufacturers of aircrafts which had accidents")
plt.show()
```
