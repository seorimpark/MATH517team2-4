# 3. Interactive plot

```python
import dash #installation of dash might be needed: 
#pip install jupyter-dash #uncomment this line to install
import dash_core_components as dcc
import dash_html_components as html
from dash.dependencies import Input, Output
import plotly.express as px
import pandas as pd

data=pd.read_csv("../Data/aircrafts_occurrences_merged.csv",encoding='latin1')

app = dash.Dash(__name__)

available_indicators = ["equipment","manufacturer","model","engine_type","engines_amount","registration_aviation","operation_phase","type_operation"]

app.layout = html.Div([
    html.Div([
        html.Div([
            dcc.Markdown("""
                **Selection of features**
                
                Choose the feature of interest to see what types of aircrafts encountered the most accidents among all.
                The plot displays a pie chart showing the shares among the different chosen categories.
                
                Note that for type of models and manufacturers, only the ones that has above 20 accidents are counted.
            """),
            dcc.RadioItems(
                id='listfeat',
                options=[{'label': i, 'value': i} for i in available_indicators],
                value="model",
                labelStyle = dict(display='block')
            )
        ], style={'width': '20%','display': 'inline-block','margin-top':'10'}),
        html.Div([
            dcc.Graph(id='indicator-graphic')
        ],style={'width': '80%','display': 'inline-block'})
    ])
])


@app.callback(
    Output('indicator-graphic', 'figure'),
    Input('listfeat', 'value'))

def update_graph(listfeat):
    #for i in listfeat:
    dat=dict(data[listfeat].value_counts())
    newd={}
    newd["names"]=list(dat.keys())
    newd["counts"]=list(dat.values())
    df=pd.DataFrame(newd)
    if listfeat=="model":
        df.loc[df['counts'] < 20, 'names'] = 'Other models'
    elif listfeat=="manufacturer":
        df.loc[df['counts'] < 20, 'names'] = 'Other manufacturers'
    fig = px.pie(df, values='counts', names='names')
    if listfeat=="engine_type":
        message="engine type"
    elif listfeat=="engines_amount":
        message="amount of engines"
    elif listfeat=="registration_aviation":
        message="registration of aviation"
    elif listfeat=="operation_phase":
        message="operation phase"
    elif listfeat=="type_operation":
        message="type of operation"
    else:
        message="type of "+listfeat
    fig.update_layout(title_text="The shares of accidents of Brazilian aircrafts coming from the same "+message+" from 2006 to 2015")
    return fig
if __name__ == '__main__':
    app.run_server(debug=False)
```
